```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
purchase_data_df.head()
```

## Player Count

* Display the total number of players


```python
SNnumber = len(purchase_data_df["SN"].unique())
total_df = pd.DataFrame({"Total Players": [SNnumber]})
total_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
NOuniqueItem = len(purchase_data_df["Item ID"].unique())
AveragePrice = round(purchase_data_df["Price"].mean())
NumofPurchase = len(purchase_data_df["Purchase ID"])
TotalRev = float(purchase_data_df["Price"].sum())



summary_df = pd.DataFrame({"Number of Unique Items": [NOuniqueItem],
                              "Average Price": AveragePrice,
                              "Number of Purchases": NumofPurchase,
                              "Total Revenue": TotalRev})

summary_df["Average Price"] = summary_df["Average Price"].map("${0:.2f}".format)
summary_df["Total Revenue"] = summary_df["Total Revenue"].map("${0:.2f}".format)
summary_df      
                       
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
Males = purchase_data_df.loc[purchase_data_df["Gender"] == "Male", :]
females = purchase_data_df.loc[purchase_data_df["Gender"] == "Female", :]
Other = purchase_data_df.loc[purchase_data_df["Gender"] == "Other / Non-Disclosed", :]

MCount = len(Males["SN"].unique())
FCount = len(females["SN"].unique())
OCount = len(Other["SN"].unique())
TCount = (MCount+FCount+OCount)

MaleP = round(MCount/TCount*100, ndigits=2) 
FemaleP = round(FCount/TCount*100, ndigits=2)
OtherP = round(OCount/TCount*100,ndigits=2)

GenderDemo_df = pd.DataFrame({
"Gender": ["Male","Female","Other/Non-Disclosed"],
'Total Count': [MCount, FCount, OCount],
"Percentage of Players" : [MaleP, FemaleP, OtherP]
})

GenderDemo_df["Percentage of Players"] = GenderDemo_df["Percentage of Players"].map("{:.2f}%".format)
GenderDemo_df.index.name = None
GenderDemo_df


```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
Males = purchase_data_df.loc[purchase_data_df["Gender"] == "Male", :]
females = purchase_data_df.loc[purchase_data_df["Gender"] == "Female", :]
Other = purchase_data_df.loc[purchase_data_df["Gender"] == "Other / Non-Disclosed", :]
#Purchase counts
MPcount = len(Males["Price"])
FPcount = len(females["Price"])
OPcount = len(Other["Price"])
#Average P Price
AMprice = round(Males["Price"].mean(),ndigits=2)
AFprice = round(females["Price"].mean(),ndigits=2)
AOprice = round(Other["Price"].mean(),ndigits=2)
#Total Purchase
TMvalue = round(Males["Price"].sum(),ndigits=2)
TFvalue = round(females["Price"].sum(),ndigits=2)
TOvalue = round(Other["Price"].sum(),ndigits=2)
#AVG Total purchase
#unique values
MCount = len(Males["SN"].unique())
FCount = len(females["SN"].unique())
OCount = len(Other["SN"].unique())
#AVG prices
ATPPM = round(TMvalue/MCount,ndigits=2)
ATPPF = round(TFvalue/FCount,ndigits=2)
ATPPO = round(TOvalue/OCount,ndigits=2)
#Create Summary table
GenderPurchase_df = pd.DataFrame({
"Gender": ["Female","Male","Other/Non-Disclosed"],
'Purchase Count': [FPcount, MPcount, OPcount],
"Average Purchase Price" : [AFprice, AMprice , AOprice],
"Total Purchase Value" : [TFvalue,TMvalue,TOvalue],
"Avg Total Purchase per Person" : [ATPPF,ATPPM,ATPPO]
})

GenderPurchase_df["Average Purchase Price"] = GenderPurchase_df["Average Purchase Price"].map("${0:.2f}".format)
GenderPurchase_df["Total Purchase Value"] = GenderPurchase_df["Total Purchase Value"].map("${0:.2f}".format)
GenderPurchase_df["Avg Total Purchase per Person"] = GenderPurchase_df["Avg Total Purchase per Person"].map("${0:.2f}".format)
GenderPurchase_df.index.name = None
GenderPurchase_df


```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#find max age
#max = purchase_data_df["Age"].max()

player_count = len(purchase_data_df["SN"].unique())
#set bins and labels
agebins = [0, 9, 14, 19, 24, 29, 34, 39, 46]
Age_Labels = ["<10", "10-14", "15-19", "20-24","25-29", "30-34", "35-39", "40+"]



#create variables for table
purchase_data_df["Age Group"] = pd.cut(purchase_data_df["Age"], bins=agebins,labels=Age_Labels)
purchase_data_df
 
age_group = purchase_data_df.groupby("Age Group")

totalcount_age = age_group["SN"].nunique()


Per_by_age = round(totalcount_age/player_count*100,2)

#create table

Demographics_df = pd.DataFrame({
   "Total Count": totalcount_age,
   "Percentage of Players": Per_by_age})


#format table
Demographics_df["Percentage of Players"] = Demographics_df["Percentage of Players"].map("{0:,.2f}%".format)
Demographics_df.index.name = None
Demographics_df




```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#set bins and labels
age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 46]
groups_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#set bins and labels

purchase_data_df["Age Group"] = pd.cut(purchase_data_df["Age"], bins=age_bins, labels=groups_names)

age_group = purchase_data_df.groupby("Age Group")

totalcount_age = age_group["SN"].nunique()

age_purchase_count = age_group["SN"].count()

avg_age_purchase_price = round(age_group["Price"].mean(),2)

total_age_purchase_value = round(age_group["Price"].sum(),2)

avg_total_age_purchase_person = round(total_age_purchase_value / totalcount_age,2)


#create table
age_purchasing_analysis_table = pd.DataFrame({
    "Purchase Count": age_purchase_count, 
    "Average Purchase Price": avg_age_purchase_price,
    "Total Purchase Value": total_age_purchase_value,
    "Avg Total Purchase per Person": avg_total_age_purchase_person
})


#format table
age_purchasing_analysis_table["Average Purchase Price"] = age_purchasing_analysis_table["Average Purchase Price"].map("${0:,.2f}".format)
age_purchasing_analysis_table["Total Purchase Value"] = age_purchasing_analysis_table["Total Purchase Value"].map("${0:,.2f}".format)
age_purchasing_analysis_table["Avg Total Purchase per Person"] = age_purchasing_analysis_table["Avg Total Purchase per Person"].map("${0:,.2f}".format)
age_purchasing_analysis_table


```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#set table variables

top_spend = purchase_data_df.groupby("SN")

top_spend_count = top_spend["Purchase ID"].count()

ave_spend_price = round(top_spend["Price"].mean(),2)

total_spend_value = top_spend["Price"].sum()

#Create Table
top_spend_table = pd.DataFrame({
    "Purchase Count": top_spend_count,
    "Average Purchase Price": ave_spend_price,
    "Total Purchase Value": total_spend_value
})

#format table
sort_top_spenders = top_spend_table.sort_values(["Total Purchase Value"], ascending=False).head()
sort_top_spenders["Average Purchase Price"] = sort_top_spenders["Average Purchase Price"].astype(float).map("${:,.2f}".format)
sort_top_spenders["Total Purchase Value"] = sort_top_spenders["Total Purchase Value"].astype(float).map("${:,.2f}".format)
sort_top_spenders










```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Define Variables to put in Summary Table

popular_item_list = purchase_data_df[["Item ID", "Item Name", "Price"]]

popular_items = popular_item_list.groupby(["Item ID","Item Name"])

purchasecount = popular_items["Price"].count()

itemprice = popular_items["Price"].sum()

purchasevalue = itemprice / purchasecount

#Create Summary DF
most_popular_items = pd.DataFrame({
   "Purchase Count": purchasecount, 
   "Item Price": purchasevalue,
   "Total Purchase Value": itemprice})

#Sort and Format table

Popular_format = most_popular_items.sort_values(["Purchase Count"], ascending=False).head()
Popular_format["Item Price"] = Popular_format["Item Price"].astype(float).map("${:,.2f}".format)
Popular_format["Total Purchase Value"] = Popular_format["Total Purchase Value"].astype(float).map("${:,.2f}".format)
Popular_format




```
